```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/koroteevmv/ML_course/main/ML4.1%20metrics/data/heart.csv")
data.head()
```

```python
y = data["output"]
x = data.drop("output", axis=1)
```

Логическая регрессия

```python
from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression(max_iter=1000).fit(x, y)
logistic.score(x, y)
```

```python
print(f"Размер датасета = {data.shape}")
```

Разделение выборки

```python
#обучающая выборка
x_train, y_train = x[:200], y[:200]
print(x_train.shape, y_train.shape)

#тестовая выборка
x_test, y_test = x[200:], y[200:]
print(x_test.shape, y_test.shape)
```

```python
logistic_test = LogisticRegression(max_iter=10000).fit(x_train, y_train)

print(f"Точность на тренировочных данных: {logistic_test.score(x_train, y_train)}")
print(f"Точность на тестовых данных: {logistic_test.score(x_test, y_test)}")
```

```python
def evaluate_logistic_regression(x, y):
    max_i, max_test = -1, -1
    train = 0

    for i in np.arange(0.6, 1.0, 0.01):
        N = int(x.shape[0] * i)
        x_train, y_train = x[:N], y[:N]
        x_test, y_test = x[N:], y[N:]

        if x_test.shape[0] == 0:
            continue

        logistic_test = LogisticRegression(max_iter=10000)
        logistic_test.fit(x_train, y_train)

        train_accuracy = logistic_test.score(x_train, y_train)
        test_accuracy = logistic_test.score(x_test, y_test)

        if test_accuracy > max_test:
            max_test = test_accuracy
            max_i = i
            train = train_accuracy

    print(f"Лучший размер тренировочного набора: {max_i * 100:.2f}% \n\nТочность на тестовых данных: {max_test:.4f} \n\nТочностью на тренировочных данных: {train_accuracy:.4f}")
```

```python
evaluate_logistic_regression(x, y)
```

```python
from numpy.random import shuffle

N = int(x.shape[0] * 0.8)

mask = np.array([True] * N + [False] * (y.shape[0] - N))
shuffle(mask)

x_train, y_train, x_test, y_test = x[mask], y[mask], x[~mask], y[~mask]
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
logistic_test = LogisticRegression(max_iter=10000).fit(x_train, y_train)

print(f"Точность на тренировочных данных: {logistic_test.score(x_train, y_train)}")
print(f"Точность на тестовых данных: {logistic_test.score(x_test, y_test)}")
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8)
x_train.shape, y_train.shape, x_test.shape, y_test.shape

logistic_test = LogisticRegression(max_iter=10000).fit(x_train, y_train)

print(f"Точность на тренировочных данных: {logistic_test.score(x_train, y_train)}")
print(f"Точность на тестовых данных: {logistic_test.score(x_test, y_test)}")
```

Построение метрик качества классификации

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
y_test_pred = logistic_test.predict(x_test)
y_train_pred = logistic_test.predict(x_train)
```

```python
print(classification_report(y_test, y_test_pred))
```

```python
print(classification_report(y_test, y_test_pred))
```

```python
metrics = pd.DataFrame({
    "Тренировочные данные": [
        accuracy_score(y_train, y_train_pred),
        precision_score(y_train, y_train_pred),
        recall_score(y_train, y_train_pred),
        f1_score(y_train, y_train_pred),
        confusion_matrix(y_train, y_train_pred)
    ],
    "Тестовые данные": [
        accuracy_score(y_test, y_test_pred),
        precision_score(y_test, y_test_pred),
        recall_score(y_test, y_test_pred),
        f1_score(y_test, y_test_pred),
        confusion_matrix(y_test, y_test_pred)
    ],
}, index=["Точность", "Прецизионность", "Полнота", "F1-меры", "Матрица путаницы"])

metrics
```
