Доп задание

```python
data_left = data_with_orders.merge(agent_data, how='left', on='Agent_name')
data_right = data_with_orders.merge(agent_data, how='right', on='Agent_name')
data_outer = data_with_orders.merge(agent_data, how='outer', on='Agent_name')
data_inner = data_with_orders.merge(agent_data, how='inner', on='Agent_name')

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

sns.heatmap(data_left.isnull(), yticklabels=False, cbar=False, ax=axes[0, 0])
axes[0, 0].set_title('Left Join')

sns.heatmap(data_right.isnull(), yticklabels=False, cbar=False, ax=axes[0, 1])
axes[0, 1].set_title('Right Join')

sns.heatmap(data_outer.isnull(), yticklabels=False, cbar=False, ax=axes[1, 0])
axes[1, 0].set_title('Outer Join')

sns.heatmap(data_inner.isnull(), yticklabels=False, cbar=False, ax=axes[1, 1])
axes[1, 1].set_title('Inner Join')

plt.tight_layout()
plt.show()
```
